5. Create and write the combined datasets

```python
#Create a version with the trimmed datasets
data = pd.concat([dataset1_expt1_trimmed, dataset1_expt2_trimmed, dataset2])
#Create the unique id for each dataset in the combined file
data['id_combined'] = (data['id'] != data['id'].shift()).cumsum()
#Delete the original id column
data = data.drop('id', axis=1)
data['id'] = data['id_combined']
data = data.drop('id_combined', axis=1)
data.head()
```

```python
#Select and order the columns to keep
data_final = data[['study', 'experiment', 'id', 'trial_num','set_size', 'change', 'response', 'rt', 'stim_size', 'fixation_size', 'stim_duration', 'retention_interval', 'distance_from_monitor', 'test_location_x_from_center', 'test_location_y_from_center']]
data_final.head()
```
